## Distribution of categories

```python
# Analyze the distribution of the 'Day' column
day_counts = df['Day'].value_counts()
print("Distribution of sales across days:\n", day_counts)

# Insights on the 'Day' column
plt.figure(figsize=(10, 6))
day_counts.plot(kind='bar')
plt.title('Distribution of Sales by Day')
plt.xlabel('Day of the Week')
plt.ylabel('Number of Sales')
plt.show()

# Calculate the percentage of sales for each day
day_percentages = (day_counts / len(df)) * 100
print("\nPercentage distribution of sales across days:\n", day_percentages)

# Identify the day with the highest sales
highest_sales_day = day_counts.idxmax()
print(f"\nThe day with the highest number of sales is: {highest_sales_day}")

# Identify the day with the lowest sales
lowest_sales_day = day_counts.idxmin()
print(f"\nThe day with the lowest number of sales is: {lowest_sales_day}")
```

```python
# Analyze the distribution of the 'Date' column
date_counts = df['Date'].value_counts()
print("Distribution of sales across dates:\n", day_counts)

# Insights on the 'Date' column
plt.figure(figsize=(10, 6))
day_counts.plot(kind='bar')
plt.title('Distribution of Sales by date')
plt.xlabel('Date of the mounth')
plt.ylabel('Number of Sales')
plt.show()

# Calculate the percentage of sales for each date
date_percentages = (date_counts / len(df)) * 100
print("\nPercentage distribution of sales across days:\n", date_percentages)

# Identify the date with the highest sales
highest_sales_date = date_counts.idxmax()
print(f"\nThe day with the highest number of sales is: {highest_sales_date}")

# Identify the date with the lowest sales
lowest_sales_date = date_counts.idxmin()
print(f"\nThe day with the lowest number of sales is: {lowest_sales_date}")
```

```python
# Analyze the distribution of 'SKU', 'BU', 'Brand', and 'Model' columns
for col in ['SKU', 'BU', 'Brand', 'Model']:
    if col in df.columns:
        counts = df[col].value_counts()
        print(f"\nDistribution of sales across {col}:\n", counts)

        plt.figure(figsize=(10, 6))
        counts.plot(kind='bar')
        plt.title(f'Distribution of Sales by {col}')
        plt.xlabel(col)
        plt.ylabel('Number of Sales')
        plt.xticks(rotation=45)
        plt.show()

        percentages = (counts / len(df)) * 100
        print(f"\nPercentage distribution of sales across {col}:\n", percentages)

        highest_sales_category = counts.idxmax()
        print(f"\nThe {col} with the highest number of sales is: {highest_sales_category}")

        lowest_sales_category = counts.idxmin()
        print(f"\nThe {col} with the lowest number of sales is: {lowest_sales_category}")
```
